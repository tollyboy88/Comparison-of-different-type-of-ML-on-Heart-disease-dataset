
# Introduction

Heart disease, also referred as cardiovascular diseases, is broad term used for diseases and conditions affecting the heart and circulatory system. It is a major cause of disability all around the world. Since heart is amongst the most vital organs of the body, its diseases affect other organs and part of the body as well. There are several different types and forms of heart diseases. The most common ones cause narrowing or blockage of the coronary arteries, malfunctioning in the valves of the heart, enlargement in the size of heart and several others leading to heart failure and heart attack."https://www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease]. This project focuses specifically on AI-driven diagnostics for heart disease, a leading cause of mortality worldwide. Bias in predicting heart disease can lead to differential treatment of patients based on non-medical factors such as race, gender, or socioeconomic status, ultimately affecting the quality and accessibility of care provided to certain demographic groups.

# About the dataset
Column Descriptions
HeartDisease : Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI).

BMI : Body Mass Index (BMI).

Smoking : Have you smoked at least 100 cigarettes in your entire life? ( The answer Yes or No ).

AlcoholDrinking : Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week

Stroke : (Ever told) (you had) a stroke?

PhysicalHealth : Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was 
your physical health not good? (0-30 days).

MentalHealth : Thinking about your mental health, for how many days during the past 30 days was your mental health not good? (0-30 days).

DiffWalking : Do you have serious difficulty walking or climbing stairs?

Sex : Are you male or female?

AgeCategory: Fourteen-level age category.

Race : Imputed race/ethnicity value.

Diabetic : (Ever told) (you had) diabetes?

PhysicalActivity : Adults who reported doing physical activity or exercise during the past 30 days other than their 14.. regular job.

GenHealth : Would you say that in general your health is...

SleepTime : On average, how many hours of sleep do you get in a 24-hour period?

Asthma : (Ever told) (you had) asthma?

KidneyDisease : Not including kidney stones, bladder infection or incontinence, were you ever told you had kidney disease?

SkinCancer : (Ever told) (you had) skin cancer?

# Data Cleaning and Exploratory Data Analysis

```python
#Preparing cleaning the agecategories column
# Drop rows with NaN values in the 'AgeCategory' column
heart_disease.dropna(subset=['AgeCategory'], inplace=True)

# Split the data by lines
lines = heart_disease['AgeCategory'].str.strip().str.split("\n")

# Remove the "Age" string from each line
cleaned_lines = []
for line in lines:
    if isinstance(line, list):
        cleaned_lines.append([item.replace("Age", "").strip() for item in line])
    else:
        cleaned_lines.append([])

# Join the cleaned lines back into a single string
cleaned_data = ["\n".join(line) for line in cleaned_lines]

# Replace the original 'AgeCategory' column with the cleaned data
heart_disease['AgeCategory'] = cleaned_data
```

# Model Development

```python
# Split data into features and target variable
X = heart_disease.drop(columns=['HadHeartAttack'])  # Features
y = heart_disease['HadHeartAttack']  # Target variable
```

# Standardization and Normanition

```python
# Standardize features (optional but recommended for some models like SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# X_test = scaler.transform(X_test)
```

```python
# Apply SMOTE to address class imbalance
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
```

```python
# initializing under and over sampling
heart_under = RandomUnderSampler(random_state=42)
heart_over = SMOTE(random_state=42)
```

```python
# under sampling of X, y 
heart_under.fit_resample(X_scaled,y)

 # over sampling of X, y
X_heart, y_heart = heart_over.fit_resample(X_scaled,y)
```

# Spltting the dataset to test and train

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_heart, y_heart, test_size=0.2, random_state=42)
```

```python
X_test = pd.DataFrame(X_test, columns=X.columns)
X_train = pd.DataFrame(X_train, columns=X.columns)
```

```python
#Display the Target variable after resampling before fit to model
labels = ['No', 'Yes']
values = y_test.value_counts().values

plt.pie(values, labels=labels, autopct='%1.0f%%')
plt.title('Heart Disease After resampling and Before the training')
```

# Model Implementation

```python
# Define the models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
}
```

# Applying GridSearchCV For Best Parameter

```python
# Parameters for GridSearchCV specific to each model
param_grids = {
    'Logistic Regression': {'classifier__C': [1]},
    'Decision Tree': {'classifier__max_depth': [None, 10]},
    'Random Forest': {'classifier__max_depth': [10], 'classifier__n_estimators': [50]},
    'Gradient Boosting': {'classifier__learning_rate': [0.1]},
    'XGBoost': {'classifier__max_depth': [5]}
}
```

# Model Evaluation

```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, classification_report, accuracy_score, precision_score
results = {}

for name, model in models.items():
    print(f"Training {name}...")
    
    pipeline = Pipeline([
        ('feature_selection', SelectKBest(f_classif, k=10)),  # Adjust k as needed
        ('classifier', model)
    ])
    
    grid_search = GridSearchCV(pipeline, param_grid=param_grids[name], cv=5, scoring='accuracy')
    
    try:
        grid_search.fit(X_train, y_train)
        y_pred = grid_search.predict(X_test)
        y_proba = grid_search.predict_proba(X_test)[:, 1]  # Assuming binary classification
        
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_proba)
        
        results[name] = {'accuracy': accuracy, 'precision': precision, 'y_proba': y_proba} # Store probabilities instead of ROC AUC score

        # Print best parameters, accuracy, precision, and classification report
        print(f"Best parameters for {name}: {grid_search.best_params_}")
        print(f"{name} Accuracy: {accuracy:.4f}")
        print(f"{name} Precision: {precision:.4f}")
        print(f"{name} ROC AUC: {roc_auc:.4f}")
        print(f"Classification Report for {name}:\n{classification_report(y_test, y_pred)}")
        
        # Confusion Matrix
        cm = confusion_matrix(y_test, y_pred)
        print(f"Confusion Matrix for {name}:\n{cm}")
        
        # Visualization of the Confusion Matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix for {name}')
        plt.xlabel('Predicted labels')
        plt.ylabel('True labels')
        plt.show()
        
        # Visualization of Prediction Distribution
        plt.figure(figsize=(8, 6))
        plt.hist(y_pred, bins=len(set(y_pred)), color='skyblue', alpha=0.75)
        plt.title(f'Prediction Distribution for {name}')
        plt.xlabel('Predicted Classes')
        plt.ylabel('Frequency')
        plt.xticks(ticks=sorted(set(y_pred)))  # Ensure one tick per class
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

    except Exception as e:
        print(f"An error occurred while training {name}: {e}")
    
    print("---------------------------------------------")
```

# visualizing the Class Variable after the prediction

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
model_names = list(results.keys())

# Accuracy
accuracies = [results[name]['accuracy'] for name in model_names]
ax[0].bar(model_names, accuracies, color='skyblue')
ax[0].set_title('Model Accuracy')
ax[0].set_ylabel('Accuracy')
ax[0].set_xticklabels(model_names, rotation=45)

# Precision
precisions = [results[name]['precision'] for name in model_names]
ax[1].bar(model_names, precisions, color='lightgreen')
ax[1].set_title('Model Precision')
ax[1].set_ylabel('Precision')
ax[1].set_xticklabels(model_names, rotation=45)

plt.tight_layout()
```

```python
plt.figure(figsize=(10, 8))

for name in model_names:
    fpr, tpr, _ = roc_curve(y_test, results[name]['y_proba'])  # Use y_proba
    roc_auc = roc_auc_score(y_test, results[name]['y_proba'])  # Calculate ROC AUC here

    plt.plot(fpr, tpr, label=f'{name} (area = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal dashed line for reference
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

# Analysis of the Result
XGBoost achieved the highest accuracy and ROC-AUC score, marking it as the most proficient model in the lineup for predicting heart disease. Its high precision rate also suggests it is reliable for predicting positive cases without many false positives.
Decision Tree also displayed excellent performance metrics, comparable to XGBoost. It excelled particularly in precision, which is critical for medical diagnostic tools where false positives can have significant implications.
Gradient Boosting showed strong overall performance, particularly in precision and ROC-AUC, making it a valuable model for scenarios where both the reliability of prediction and the ability to differentiate between classes are crucial.
Random Forest and Logistic Regression showed good performance but were slightly outpaced by the ensemble methods. While not as effective as XGBoost or Decision Tree in this context, they still hold substantial predictive power and might be preferred in scenarios requiring less computational intensity or greater interpretability.
The ROC-AUC scores were notably high for all models, especially XGBoost, indicating excellent capability across these models to distinguish between the patient having or not having heart disease across various threshold settings.
